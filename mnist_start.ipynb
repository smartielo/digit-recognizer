```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# ðŸ‡§ðŸ‡· Baixando o dataset (pode levar alguns segundos)
# ðŸ‡ºðŸ‡¸ Downloading the dataset (might take a few seconds)
print("ðŸ‡§ðŸ‡· Baixando dados... / ðŸ‡ºðŸ‡¸ Downloading data...")
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='auto')

# ðŸ‡§ðŸ‡· Separando em X (Imagens/Pixels) e y (Respostas/Labels)
# ðŸ‡ºðŸ‡¸ Splitting into X (Images/Pixels) and y (Answers/Labels)
X, y = mnist["data"], mnist["target"]

print(f"Shape X: {X.shape}")
print(f"Shape y: {y.shape}")
```

```python
# ðŸ‡§ðŸ‡· Pegar o primeiro dÃ­gito do dataset (Ã­ndice 0)
# ðŸ‡ºðŸ‡¸ Get the first digit of the dataset (index 0)
some_digit = X[0]

# ðŸ‡§ðŸ‡· Transformar a linha de 784 de volta num quadrado 28x28
# ðŸ‡ºðŸ‡¸ Reshape the 784 line back into a 28x28 square
some_digit_image = some_digit.reshape(28, 28)

# ðŸ‡§ðŸ‡· Plotar a imagem em escala de cinza
# ðŸ‡ºðŸ‡¸ Plot the image in grayscale
plt.imshow(some_digit_image, cmap="binary")
plt.axis("off") # Remove a rÃ©gua/bordas
plt.show()

# ðŸ‡§ðŸ‡· O que o gabarito diz?
# ðŸ‡ºðŸ‡¸ What does the label say?
print(f"Label: {y[0]}")
```
